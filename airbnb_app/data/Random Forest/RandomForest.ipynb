```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

data_abb = pd.read_csv('clean_listings.csv')
```

```python
data_abb = data_abb.rename(columns={'neighbourhood_cleansed': 'neighbourhood_group', 'neighbourhood_group_cleansed': 'neighbourhood'})
data_abb.head()
```

```python
# data_abb = data_abb.rename(columns={'neighbourhood_cleansed': 'neighbourhood', 'neighbourhood_group_cleansed': 'neighbourhood'}
```

```python
df=data_abb[['price','neighbourhood_group','minimum_nights','room_type',
             'availability_365', 'number_of_reviews']]
df.head()
```

```python
df_dummies = pd.get_dummies(df,prefix=['neighbourhood_group', 'room_type'])
df_dummies.to_csv('ABB_ML_ready.csv')
df_dummies.head()
```

```python
X = df_dummies.drop('price', axis=1)

y = df_dummies['price'].values.reshape(-1,1)
# X dependent, y independent  
# X is one variable, raising by a degree each time, 
# y is the resultant;
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python

regressor.score(X_test, y_test)
```
